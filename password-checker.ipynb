```python
from re import U
password=input("Enter a password")
score=0
if len(password)>=8:
  score+=1

if len(password)<=7:
  print("longer password required")

uppercase_list = [chr(i) for i in range(65, 91)]

lowercase_list = [chr(i) for i in range(97, 123)]

# Check if the password contains any uppercase and or lowercase letters
if not any(char in uppercase_list for char in password):
  print("no uppercase letters")

if not any(char in lowercase_list for char in password):
  print("no lowercase letters")

import string

# This contains: "0123456789"
numbers_code = string.digits

if not any(char in numbers_code for char in password):
 print("no numbers")

# This contains all standard special characters
symbols_code = string.punctuation

if not any(char in symbols_code for char in password):
  print("missing symbols")

# Corrected scoring logic
if any(char in uppercase_list for char in password):
  score+=1

if any(char in lowercase_list for char in password):
 score+=1

if any(char in numbers_code for char in password):
  score+=1

if any(char in symbols_code for char in password):
  score+=1

if len(password)>=12:
  score+=1

print(score)

if score>=5: print("strong password")
elif score>=3:print("medium password")
else:print("weak password")
```
